# November 2024 Puzzle Solution

You’re hosting a 30-person Thanksgiving and need someone to bring a can of cranberry sauce. For each person you ask, the probability that they end up bringing a Can o’ Cran is 0.1. If you ask all 30 people, what’s the probability you’ll get exactly one can?

Bonus: You need cranberry sauce, but you don’t want to get bogged down with a lot of it. How many people should you ask to maximize the chance that you get between 1 and 3 cans?


Reference solution in [brilliant](https://brilliantorg.notion.site/November-puzzle-solution-13fb4926c4858050a3acc3a4ad2630db)

```python
p = 0.1    # Probability that a person bringing a Can o’ Cran
q = 1 - p  # Probability that a person does not bring a Can o’ Cran
m = 30     # numner of persons coming on Thanksgiving
```

```python
P_exactly_1 = m*p*q**(m-1)
f'The probability I’ll get exactly one can: {P_exactly_1*100:.1f}%'
```

```python
def P_1_to_3(n):
    '''
    Getting between 1 and 3 cans means getting either 1 or 2 or 3.
    Using binomial coefficients after simplification we have 
    '''
    return n*p*q**(n-3) * (q**2  + (n-1)/2*p*q + (n-1)*(n-2)/6*p**2)
```

```python
'''
Plot probability as a function  of n we may find n which corresponds to the function maximum
'''

import numpy as np
import matplotlib.pyplot as plt

n_values = np.arange(1, m+1)  # n from 1 to 30
p_1_to_3_values = [P_1_to_3(n) for n in n_values]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(n_values, p_1_to_3_values, color='blue', label='P_1_to_3(n)', s=10)  # Adjust size for better visibility
plt.title('Scatter Plot of P_1_to_3(n) with Interstices')
plt.xlabel('n')
plt.ylabel('P_1_to_3(n)')
plt.grid(True)
plt.legend()
plt.show()

```
